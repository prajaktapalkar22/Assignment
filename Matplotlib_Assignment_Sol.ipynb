```python
# Matplotlib #
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
df.head()
```

```python
# Example 1 - a pie chart presenting the male/female proportion 
labels=['Male','Female']
# Creating plot 
fig = plt.figure(figsize =(10, 7)) 

plt.pie(df['sex'].value_counts(), labels = labels, autopct='%1.1f%%') 

# show plot 
plt.show() 
```

```python
# Example 2 - a scatterplot with the Fare paid and the Age
groups = df.groupby('sex')
for name, group in groups:
    plt.scatter(group.fare, group.age, label=name)

plt.legend()
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Scatter Plot')
```
