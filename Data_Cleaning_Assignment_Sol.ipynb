```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'], 
 'FlightNumber': [10045, np.nan, 10065, np.nan, 10085], 
 'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]], 
 'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )', 
 '12. Air France', '"Swiss Air"']})
                  
```

```python
# 1. Add missing values in Flight Number Column
lst=[]
lst=df.loc[df['FlightNumber'].isnull()==True].index.astype(list)
for i in lst:
    df.loc[i,'FlightNumber']=df.loc[i-1,'FlightNumber']+10
            
# Change datatype to int
df['FlightNumber'] = df['FlightNumber'].astype(int) 
df
```

```python
# 2. Update From_To column as two separate columns
fromlst=[]
tolst=[]
for i in df['From_To']:
    lst=i.split('_')
    fromlst.append(lst[0])
    tolst.append(lst[1])

df1=pd.DataFrame(columns =['From', 'To'])
df1['From']=fromlst
df1['To']=tolst
df1
```

```python
# 3. Capitalisation of the city names 
df1=df1.applymap(lambda x : x.capitalize())
df1
```

```python
# 4. Delete the From_To column from df and attach the temporary DataFrame 
df.drop('From_To',  axis=1, inplace=True)
df = pd.concat([df1, df], axis=1)
df
```

```python
# 5. Expand RecentDelays Column and replace with expanded data in original data frame
delays = df['RecentDelays'].apply(pd.Series)

delays = delays.rename(columns = lambda x : 'delay_' + str(x))

delays

df=pd.concat([df[['From','To','FlightNumber']],delays,df[['Airline']]], axis=1)
df
```
